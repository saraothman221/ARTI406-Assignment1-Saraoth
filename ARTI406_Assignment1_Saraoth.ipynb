## 4. Data Visualization
Visualizing the relationships between Age, Annual Income, and Spending Score to identify patterns.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df , color="yellow")
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop(['CustomerID', 'Gender'], axis=1).corr(), annot=True, cmap='RdYlGn')
plt.title('Correlation Heatmap')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'],bins=20,kde=True,color='green')
plt.title("----Distrbution of customer age ----")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', s=100)
plt.title('Income vs Spending Score (Colored by Gender)')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
plt.title('Checking for Outliers')
plt.show()
```

## 5. Conclusion and Insights
