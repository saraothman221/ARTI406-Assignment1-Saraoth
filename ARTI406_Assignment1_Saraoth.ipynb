## 4. Data Visualization
Visualizing the relationships between Age, Annual Income, and Spending Score to identify patterns.

Why: To identify the gender balance of the mall customers and see which group visits more.
Insight: There are more female customers (56%) than male customers (44%), suggesting that marketing campaigns should consider female preferences.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', data=df , color="yellow")
plt.title('Gender Distribution')
plt.show()
```

Why: To check the strength of the relationship between numerical features (Age, Income, Spending Score).
Insight: There is a negative correlation between Age and Spending Score, meaning younger customers generally tend to spend more.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop(['CustomerID', 'Gender'], axis=1).corr(), annot=True, cmap='RdYlGn')
plt.title('Correlation Heatmap')
plt.show()
```

Why: To determine the most common age group among customers for targeted advertising.
Insight: The majority of customers are in their 30s, with a significant drop in customers older than 50.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'],bins=20,kde=True,color='green')
plt.title("----Distrbution of customer age ----")
plt.show()
```

Why: To get a comprehensive view of all numerical relationships and distributions in one grid.
Insight: The pairplot confirms that Spending Score is the most helpful feature for identifying different customer segments.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Gender', s=100)
plt.title('Income vs Spending Score (Colored by Gender)')
plt.grid(True)
plt.show()
```

Why: To visualize the distribution of spending scores, identify the median, and check for any potential outliers (unusual data points).

Insight: The box plot shows that the spending scores are consistently distributed between 1 and 100 with a median around 50. There are no outliers in the spending score, which indicates that the data is stable and ready for modeling.

```python
plt.figure(figsize=(10, 5))
sns.boxplot(data=df[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']])
plt.title('Checking for Outliers')
plt.show()
```

Why: To understand the income distribution and purchasing power of our customers.
Insight: Most customers earn between $50k and $80k, showing a strong middle-class presence

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Annual Income (k$)'], bins=20, kde=True, color='green')
plt.title('Distribution of Annual Income')
plt.show()
```

Why: To identify the spread of spending scores and detect any potential outliers.
Insight: The spending scores are well-distributed between 1 and 100, with a median around 50 points.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Spending Score (1-100)'], color='gold')
plt.title('Spending Score Range and Outliers')
plt.show()
```

Why: To compare the average spending behavior between Male and Female customers.
Insight: Female customers have a slightly higher average spending score compared to male customer

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Gender', y='Spending Score (1-100)', data=df, palette='magma')
plt.title('Average Spending Score: Male vs Female')
plt.show()
```
